```python
#divide_image = sub_image/rms_image
#Load the working image 

divide_image = np.load(r"C:\Users\jacob\AstronomyResearchStuff\WorkingImage.npy")


# find contours at 3 sigma
contours = measure.find_contours(divide_image, 3)

# save vertices, where each key in the dictionary is a key to a unique x,y vertices stack
verts = {}

for index in range(0,len(contours)):
    verts[index] = np.stack((contours[index][:,1],contours[index][:,0]),axis=-1)

# save dictionary to file
with open('vertices_all3sigma.pkl', 'wb') as fp:
    pickle.dump(verts, fp)
    print('dictionary saved successfully to file')

# next, open dictionary and get radius, center, and area:

with open('vertices_all3sigma.pkl', 'rb') as fp:
    verts3 = pickle.load(fp)

radius3 = {}
center3 = {}

contourInfo = {}
contourInfo['key'] = []
contourInfo['radius'] = []
contourInfo['center'] = []
contourInfo['area'] = []

radius3_array = []
center3_array = []

for key in verts3.keys():

    vertices = verts3[key]

    x = [i[0] for i in vertices]
    y = [i[1] for i in vertices]

    polygon = Polygon(vertices)

    area1 = polygon.area
    radius1 = np.sqrt(area1/np.pi)
    radius3[key]=radius1
    center1 = list(polygon.centroid.coords)[0]
    center3[key]=center1
    radius3_array.append(radius1)
    center3_array.append(center1)

    #Save the contour information to a dictionary
    contourInfo['key'].append(key)
    contourInfo['radius'].append(radius1)
    contourInfo['center'].append(center1)
    contourInfo['area'].append(area1)

#Save the contour information to an astropy table

contour_table = Table(contourInfo)

#Write the table to an fits file

contour_table.write('ContourInfo.fits', format = 'fits')
```
